```python
import tensorflow as tf
import nltk
from nltk.corpus import stopwords
import re

# stpwrds = stopwords.words("english")
# full_text = text

def preprocess(X_batch):
    p_result = []
    i = 0
    for review in X_batch:
        review = review[0]
        if i != 10:
            print("Original: ", review)
        review = re.sub(r"<br\s*/?>", " ", review)
        review = re.sub("[^a-zA-Z']", " ", review)
        tokens = nltk.word_tokenize(review)
        tokens = tokens[:300]
#         tokens = [token for token in tokens if token not in stpwrds]
        review = " ".join(token for token in tokens)
        if i != 10:
            print("Preprocessed: ", review)
        i += 1
        p_result.append(review)
        
    return p_result
```
