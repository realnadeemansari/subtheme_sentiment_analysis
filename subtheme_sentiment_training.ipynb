```python
dict_map = {'advisoragent service':'advisor/agent service', 'tyre agedot code':'tyre age/dot code'}    

def map_contractions(s):
    if s in dict_map.keys():
        return dict_map[s]
    else:
        return s
```

```python
dict_map.keys()
```

```python
def extract_pos_neg(s):
    if 'positive' in s:
        return (s, 1)
    else:
        return (s, 1)
```

```python
for i, row in y_data.iterrows():
#     print(row.iloc[0])
    lis_subthemes = row.dropna().tolist()
#     print('\n', lis_subthemes)
    final_subthemes = list(map(map_contractions, lis_subthemes))
    
    final_subthemes2 = list(map(extract_pos_neg, final_subthemes))
    
    for t in final_subthemes2:
        data1.loc[i][t[0]] = t[1]
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow import keras
```

```python
def make_model():
    model = keras.Sequential([
        layers.Dense(1025, activation='relu', input_shape=[X_train.shape[1]]),
        layers.Dropout(0.3),
        layers.Dense(512, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(128,activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(len(s), activation='sigmoid'),
    ])
    
    return model
```

```python
model = make_model()
model.summary()
```

```python
early_stopping_cb = keras.callbacks.EarlyStopping(patience=5)
model.compile(optimizer=keras.optimizers.RMSprop(1e-3), loss="binary_crossentropy", metrics=["accuracy"])
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=32, callbacks=[early_stopping_cb], validation_split=0.2)
```
